```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.simplefilter('ignore')
```

```python
data=pd.read_csv("Raw_Housing_Prices.csv")
data.head()
```

```python
data.describe()
```

Target variable is Sale Price

```python
data['Sale Price'].plot.hist()
```

```python
#Checking quantile
q1=data['Sale Price'].quantile(0.25)
q3=data['Sale Price'].quantile(0.75)
q1,q3
```

```python
#interquartile range
iqr=q3-q1
iqr
```

```python
upper_limit=q3+1.5*iqr
lower_limit=q1-1.5*iqr
upper_limit,lower_limit
```

```python
#imputing outlier
def limit_imputer(value):
    if(value>upper_limit):
        return upper_limit
    if(value<lower_limit):
        return lower_limit
    else:
        return value
```

```python
#using apply func
data['Sale Price']=data['Sale Price'].apply(limit_imputer)
```

```python
#Let see the plot after outlier treatment
data['Sale Price'].plot.hist()
```

```python
#checking if any column value is null or not
data.isnull().sum()
```

```python
#removing target variable if it is 
data['Sale Price'].dropna(inplace=True)
data['Sale Price'].isnull().sum()
```

```python
data.info()
```

```python
#isolating numerical column
numerical_columns=['No of Bathrooms','Flat Area (in Sqft)','Lot Area (in Sqft)','Area of the House from Basement (in Sqft)',
'Latitude','Longitude','Living Area after Renovation (in Sqft)']
```

```python
#imputing missing values for numerical columns
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='median')
data[numerical_columns]=imputer.fit_transform(data[numerical_columns])
```

```python
data.info()
```

```python
#imputing zipcode
#we have to reshape to 2 dimensional array
imputer=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
data['Zipcode']=imputer.fit_transform(data['Zipcode'].values.reshape(-1,1))
```

```python
data.info()
```

```python
data['No of Times Visited'].unique()
```

```python
mapping={'None':"0",
         'Once':"1",
         'Twice':"2",
         'Thrice':"3",
         'Four':"4"}

data['No of Times Visited']=data['No of Times Visited'].map(mapping)
```

```python
data['No of Times Visited'].unique()
```

```python
data['Ever Renovated']=np.where(data['Renovated Year']==0,'No','Yes')
```

```python
#manipuldating date time variable
data['Purchase Year']=pd.DatetimeIndex(data['Date House was Sold']).year
data['Purchase Year']
```

```python
#finding how many years ago the house was renovated
data['Year Since Renovation']=np.where(data['Ever Renovated']=='Yes',abs(data['Purchase Year']-data['Renovated Year']),0)
```

```python
data.head()
```

```python
#dropping redundant column
data.drop(columns=['Purchase Year','Date House was Sold','Renovated Year'])
```

```python
#Checking 'Condition of the house' with sale price mean
data.groupby('Condition of the House')['Sale Price'].mean().sort_values().plot(kind='bar')
```

```python
#Checking 'Waterfront View' with sale price mean
data.groupby('Waterfront View')['Sale Price'].mean().sort_values().plot(kind='bar')
```

```python
#Checking 'Ever Renovated' with sale price mean
data.groupby('Ever Renovated')['Sale Price'].mean().sort_values().plot(kind='bar')
```
